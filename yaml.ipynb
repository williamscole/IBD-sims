```python
def find_directory():
    with open("simulation_runs.pkl", "rb") as pklf:
        d = pkl.load(pklf)
        print(d)
    # del config["label"]
    
    for i,j in d.items():
        print(i, j)
        if j == config:
            return i
    return "Not found!"

find_directory()
```
